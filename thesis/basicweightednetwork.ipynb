```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
k = 1
limit = 10
while (k < limit) :
    G.add_node(k, b1 = np.random.random(), b2 = np.random.random())
    k += 1
while (k > 1):
    x = np.random.randint(1,limit)
    y = np.random.randint(1,limit)
    G.add_edge(x,y, weight = G.node[x]['b1'] - G.node[y]['b1'])
    k -= 1
```

```python
elarge=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] >0.5]
esmall=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] <=0.5]

pos=nx.spring_layout(G)
```

```python
nx.draw_networkx_nodes(G,pos,node_size=100, node_color='g')

nx.draw_networkx_edges(G,pos,edgelist=elarge,
                    width=6)
nx.draw_networkx_edges(G,pos,edgelist=esmall,
                    width=6,alpha=0.5,edge_color='b',style='dashed')

nx.draw_networkx_labels(G,pos,font_size=2,font_family='sans-serif')
plt.show()
```
